```python
import pandas as pd
df3 = pd.read_csv("heart-disease.csv")
df3
```

```python
df3.age.plot.hist(bins=10); 
```

```python
df3.plot.hist(figsize = (10,30), subplots = True)
```

```python
over50 = df3[df3['age']>50]
over50 = over50.reset_index()
over50.drop('index', axis = 1, inplace = True)
```

```python
over50
```

```python
import matplotlib.pyplot as plt 
over50.plot(kind = 'scatter', x = 'age', y = 'chol', c = "target");
```

```python
fix, ax = plt.subplots(figsize = (20,9))
over50.plot(kind = "scatter", x = 'age', y = 'chol', c = 'target', ax=ax);

ax.set_xlim([45,100])
```

```python
over50.plot(kind = "scatter", x = 'age', y = 'chol', c = 'target', figsize = (20,9));
```

```python
fig, ax = plt.subplots(figsize = (10,6))
scatter = ax.scatter(x =over50['age'] , y = over50["chol"], c = over50['target'])
ax.set(title = 'Heart Disease and Cholestrol Levels',
      xlabel='Age',
      ylabel='Cholestrol')

ax.legend(*scatter.legend_elements(), title='target')
ax.axhline(y=over50['chol'].mean(), linestyle = "-.")
```

```python
over50.head()
```

```python
fig, (ax0, ax1) = plt.subplots(nrows = 2, ncols = 1, figsize = (10,10), sharex=True)

scatter = ax0.scatter(x = over50['age'], y = over50['chol'], c = over50['target'])
ax0.set(title = 'Heart disease and Chol Level',
       ylabel = 'cholestrol')

ax0.legend(*scatter.legend_elements(), title = 'target')
ax0.axhline(y=over50['chol'].mean(), linestyle = '-.')



ax1.scatter(x = over50['age'], y = over50['thalach'], c = over50['target'])
ax1.set(title = 'Heart disease and Thalach Level',
       xlabel = 'age',
       ylabel = 'thalach')

ax1.legend(*scatter.legend_elements(), title = 'target')
ax1.axhline(y=over50['thalach'].mean(), linestyle = '-.')

fig.suptitle('Heart Disease Analysis', fontsize = 16, fontweight='bold');
```
